```python
tsource = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
tsoup=BeautifulSoup(tsource,'lxml')
print(tsoup.title)
from IPython.display import display_html
tab = str(tsoup.table)
display_html(tab,raw=True)
```

```python
dfst = pd.read_html(tab)
dft=dfst[0]
dft.head()

```

```python
# Drop rows where Borough is 'Not assigned'
dft1 = dft[dft.Borough != 'Not assigned']

# Combine the neighbourhoods with same Postal code
dft2 = dft1.groupby(['Postcode','Borough'], sort=False).agg(', '.join)
dft2.reset_index(inplace=True)

# Replacing the name of the neighbourhoods which are 'Not assigned' with names of Borough
dft2['Neighborhood'] = np.where(dft2['Neighborhood'] == 'Not assigned',dft2['Borough'], dft2['Neighborhood'])

dft2
```
